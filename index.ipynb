```python
import pandas as pd
import os
from pathlib import Path

ads_links =pd.read_csv("./ads.txt")
ai_links =pd.read_csv("./ai.txt")
dac_links=pd.read_csv("./dac.txt")
print(ads_links,ai_links,dac_links)
```

```python
ads_links.drop(index=8,inplace=True)
ads_links.reset_index(drop=True,inplace=True)
ads_links
```

```python
dac_links.drop(index=[0,1],inplace=True)
dac_links.reset_index(drop=True,inplace=True)
dac_links
```

```python
dataset=pd.concat([ads_links,ai_links,dac_links])

dataset.reset_index(drop=True,inplace=True)
dataset.drop_duplicates()
```

```python
dataset.to_json('./dataset.json',orient='records')
```

```python

for index, row in dataset.iterrows():
    os.makedirs('d:\\dataset\\datasets\\'+row['Dataset_Name'])
    dataset.at[index, 'path'] ='d:\\dataset\\datasets\\'+row['Dataset_Name']
    
```
